```python
def fibonacci(n):
    """Generate the Fibonacci sequence up to n terms."""
    sequence = []
    a, b = 0, 1
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence
num_terms = 10
fib_sequence = fibonacci(num_terms)
print(f"Fibonacci sequence up to {num_terms} terms: {fib_sequence}")
```

```python
def prime_factors(n):
    """Return the prime factors of a given positive integer."""
    factors = []
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n //= i
    if n > 2:
        factors.append(n)
    return factors
number = 673456
factors = prime_factors(number)
print(f"Prime factors of {number}: {factors}")
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6  
    return True
print(is_prime(29))
```

```python
def fibonacci_sequence(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    fib_sequence = [0, 1]
    for i in range(2, n):
        next_term = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_term)
    return fib_sequence
print(fibonacci_sequence(10))
```

```python
def prime_factors(n):
    factors = []
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    factor = 3
    while factor * factor <= n:
        while n % factor == 0:
            factors.append(factor)
            n //= factor
        factor += 2
    if n > 2:
        factors.append(n)
    return factors
print(prime_factors(56))  
print(prime_factors(101))
print(prime_factors(45))
print(prime_factors(1))
```

```python
def most_common_word(paragraph):
    paragraph = ''.join(char if char.isalnum() or char.isspace() else ' ' for char in paragraph.lower())
    words = paragraph.split()
    word_counts = {}
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    most_common = max(word_counts, key=word_counts.get)
    return most_common, word_counts[most_common]
paragraph = "hi hi hi hi hi hi hi hi hi"
word, count = most_common_word(paragraph)
print(f"The most common word is: '{word}' with {count} occurrences.")
```

```python
class Student:
    def __init__(self, student_id, name):
        self.id = student_id
        self.name = name
        self.grades = []
    def add_grade(self, grade):
        self.grades.append(grade)
    def avg_grade(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0
students = {}
def add_student(student_id, name):
    students[student_id] = Student(student_id, name)
def assign_grade(student_id, grade):
    students.get(student_id, Student(None, None)).add_grade(grade)
def display_student(student_id):
    s = students.get(student_id)
    if s:
        print(f"ID: {s.id}, Name: {s.name}, Avg Grade: {s.avg_grade():.2f}")
add_student(1, "Vikrant")
add_student(2, "Petluri")
assign_grade(1, 100)
assign_grade(1, 90)
assign_grade(2, 78)
display_student(1)
display_student(2)
```

```python
import cv2
import numpy as np
import time
from ultralytics import YOLO
import math

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("C:/Users/vikra/Downloads/56310-479197605_small.mp4")
fps = cap.get(cv2.CAP_PROP_FPS)

object_count = 0
object_ids_crossed = set()
left_street_count = 0
right_street_count = 0
object_positions = {}

METERS_PER_PIXEL_TOP = 5 / 100     
METERS_PER_PIXEL_MID = 6 / 100     
METERS_PER_PIXEL_BOTTOM = 7 / 100  

LANE_WIDTH_METERS = 3.7  # One lane width
LANE_PIXEL_WIDTH = 250   # Adjust this based on actual video measurement if needed

if not cap.isOpened():
    print("Error: Could not open video.")
else:
    print("Streaming from video... Press 'q' to quit.")
    try:
        while True:
            ret, color_frame = cap.read()
            if not ret:
                print("Failed to grab frame")
                break

            color_image = np.asanyarray(color_frame)
            height, width, _ = color_image.shape
            line_y = height // 2
            line_x = width // 2

            results = model.track(source=color_image, persist=True)[0]
            label_image = results.plot()

            cv2.line(label_image, (0, height // 3), (width, height // 3), (200, 200, 200), 1)
            cv2.line(label_image, (0, 2 * height // 3), (width, 2 * height // 3), (200, 200, 200), 1)
            cv2.line(label_image, (line_x, 0), (line_x, height), (255, 0, 0), 2)
            cv2.line(label_image, (0, line_y), (width, line_y), (0, 255, 255), 2)

            if results.boxes is not None:
                for box in results.boxes:
                    x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())
                    center_x = (x1 + x2) / 2
                    center_y = (y1 + y2) / 2

                    if box.id is not None:
                        obj_id = int(box.id[0])
                        current_time = time.time()

                        if obj_id in object_positions:
                            prev_x, prev_y, prev_time = object_positions[obj_id]
                            tt = current_time - prev_time

                            if tt > 0:
                                dx = center_x - prev_x
                                dy = center_y - prev_y
                                pixel_distance = math.hypot(dx, dy)

                                # Perspective scaling
                                if center_y < height // 3:
                                    meters_per_pixel = METERS_PER_PIXEL_TOP
                                elif center_y < 2 * height // 3:
                                    meters_per_pixel = METERS_PER_PIXEL_MID
                                else:
                                    meters_per_pixel = METERS_PER_PIXEL_BOTTOM

                                distance_center = pixel_distance * meters_per_pixel
                                speed_center = distance_center / tt

                                # Lane-based X movement
                                meters_per_pixel_lane = LANE_WIDTH_METERS / LANE_PIXEL_WIDTH
                                distance_lane = abs(dx) * meters_per_pixel_lane
                                speed_lane = distance_lane / tt

                                speed_mps = (speed_center + speed_lane) / 2

                                label_text = f"Speed: {speed_mps:.2f} m/s"
                                cv2.putText(label_image, label_text, (x1, y1 + 35),
                                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2)

                        object_positions[obj_id] = (center_x, center_y, current_time)

                        if center_y > line_y and center_x < line_x:
                            if obj_id not in object_ids_crossed:
                                object_ids_crossed.add(obj_id)
                                object_count += 1
                                left_street_count += 1

                        if center_y < line_y and center_x > line_x:
                            if obj_id not in object_ids_crossed:
                                object_ids_crossed.add(obj_id)
                                object_count += 1
                                right_street_count += 1

            cv2.putText(label_image, f'Objects passed: {object_count}', (10, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.putText(label_image, f'Left Street: {left_street_count}', (10, 80),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
            cv2.putText(label_image, f'Right Street: {right_street_count}', (10, 120),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

            cv2.imshow("YOLOv8 Detection", label_image)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        cap.release()
        cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("C:/Users/vikra/Downloads/ytdl.canehill.info - Vehicle Flow (720p).mp4")

line_y1 = 240
line_y2 = 520
real_distance_meters = 66.66

fps = cap.get(cv2.CAP_PROP_FPS)
print(fps)
frame_number = 0

object_frames = {}
object_speeds = {}
object_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_number += 1
    height, width, _ = frame.shape
    results = model.track(source=frame, persist=True)[0]
    output_frame = results.plot()

    if results.boxes is not None:
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())
            center_y = int((y1 + y2) / 2)

            if hasattr(box, 'id') and box.id is not None:
                obj_id = int(box.id[0])

                if obj_id not in object_frames:
                    object_frames[obj_id] = {}

                if 'start' not in object_frames[obj_id] and center_y > line_y1:
                    object_frames[obj_id]['start'] = frame_number

                if 'start' in object_frames[obj_id] and 'end' not in object_frames[obj_id] and center_y > line_y2:
                    object_frames[obj_id]['end'] = frame_number
                    frame_diff = object_frames[obj_id]['end'] - object_frames[obj_id]['start']
                    if frame_diff > 0:
                        speed_kph = ((real_distance_meters * fps) / frame_diff) * 3.6
                        object_speeds[obj_id] = speed_kph
                        object_count += 1

                if obj_id in object_speeds:
                    speed = object_speeds[obj_id]
                    cv2.putText(output_frame, f"{speed:.2f} km/h", (x1, y1 + 35),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2)

    cv2.line(output_frame, (0, line_y1), (width, line_y1), (0, 255, 0), 2)
    cv2.line(output_frame, (0, line_y2), (width, line_y2), (0, 0, 255), 2)

    cv2.putText(output_frame, f'Objects measured: {object_count}', (10, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

    cv2.imshow("YOLOv8 Speed Estimation", output_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pyrealsense2 as rs
import numpy as np
import cv2
from ultralytics import YOLO
import time

pipeline = rs.pipeline()
config = rs.config()
config.enable_stream(rs.stream.depth, 640, 480, rs.format.z16, 30)
config.enable_stream(rs.stream.color, 640, 480, rs.format.bgr8, 30)
pipeline.start(config)
cv2.namedWindow('D455 AI Detection', cv2.WINDOW_AUTOSIZE)

model = YOLO("C:/Users/vikra/Downloads/BallsYOLOR.v15i.yolov8/runs/detect/train3/weights/best.pt")


object_history = {}

try:
    while True:
        frames = pipeline.wait_for_frames()
        depth_frame = frames.get_depth_frame()
        color_frame = frames.get_color_frame()
        if not depth_frame or not color_frame:
            continue

        depth_image = np.asanyarray(depth_frame.get_data())
        color_image = np.asanyarray(color_frame.get_data())
        rgb_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

        results = model.track(source=color_image,persist=True,conf=0.3,iou=0.5)

        if results[0].boxes is not None:
            for box in results[0].boxes:
                if box.id is None:
                    continue

                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = float(box.conf[0])
                cls = int(box.cls[0])
                obj_id = int(box.id) if box.id is not None else -1
                label = model.names[cls]

                center_x = int((x1 + x2) / 2)
                center_y = int((y1 + y2) / 2)
                curr_time = time.time()

                speed_text = "Speed: -- px/s"
                if obj_id in object_history:
                    prev_x, prev_y, prev_time = object_history[obj_id]
                    at = curr_time - prev_time
                    ax = center_x - prev_x
                    ay = center_y - prev_y
                    dist = np.sqrt(ax**2 + ay**2)
                    if at > 0:
                        speed = dist / at
                        speed_text = f"Speed: {speed:.2f} px/s"

                object_history[obj_id] = (center_x, center_y, curr_time)

                text = f'{label} {conf:.2f} {obj_id}'
                cv2.rectangle(color_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(color_image, text, (x1, y1 - 25),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
                cv2.putText(color_image, speed_text, (x1, y1 - 5),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
            

        depth_colormap = cv2.applyColorMap(
            cv2.convertScaleAbs(depth_image, alpha=0.03),
            cv2.COLORMAP_JET
        )
        depth_colormap_resized = cv2.resize(depth_colormap, (color_image.shape[1], color_image.shape[0]))
        combined = np.hstack((color_image, depth_colormap_resized))

        cv2.imshow('D455 AI Detection', combined)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

finally:
    pipeline.stop()
    cv2.destroyAllWindows()
```
